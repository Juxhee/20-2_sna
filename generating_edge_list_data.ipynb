```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize 
```

```python
stop = ['.',';',',',')','The',"'s",'We','in','us','A','1','2','3','4','(',':','"',"'","''",'--','%','\n',
       '△ Less','Less','△','paper','framework','model','models','data','methods','approach','input','approaches',
       'task','use','result','results','dataset','present','presents','several','propose','tasks']
```

```python
stop_words = set(stopwords.words('english') + stop)
```

```python
data = pd.read_csv('./arxiv_data.csv')
```

```python
data = data.drop(['Unnamed: 0'],axis=1)
```

```python
data.columns=['title','abstract']
```

```python
def get_nn(x):
    
    nouns = word_tokenize(x)
    nouns = [noun for noun in nouns if len(noun)>1]
    nouns = [noun for noun in nouns if noun not in stop_words]
    nouns = nltk.pos_tag(nouns)
    nn_words = []
    for word, pos in nouns : 
        if pos in ['NN', 'NNS', 'NNP', 'NNPS', 'JJ', 'JJR', 'JJS']:
            nn_words.append(word)
    return nn_words
```

```python
title = data['title'].apply(lambda x : get_nn(x))
```

```python
def get_comb(x):
    a = list(combinations(x,2))
    return a
```

```python
# 같이 등장한 단어 조합 만들기 
title = title.apply(lambda x : get_comb(x))
```
